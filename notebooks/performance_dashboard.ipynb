```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

df = pd.read_csv("../results/strategy_comparison.csv")
df.head()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = df, x="strategy", y="return_pct", hue="symbol")
plt.title("Net Return % per Strategy / Symbol")
plt.ylabel("Net Return (%)")
plt.grid(True, axis="y", ls="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
fig = px.scatter(
    df,
    x="max_drawdown",
    y="return_pct",
    color="strategy",
    symbol="symbol",
    hover_data=["total_trades", "win_rate"],
    title="Return vs Max Drawdown (interactive)"
)
fig.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data = df, x="strategy", y="total_trades", hue="symbol")
plt.title("Total Trades per Strategy / Symbol")
plt.ylabel("Total Trades")
plt.grid(True, axis="y", ls="--", alpha=0.7)
plt.tight_layout()
plt.show()
```
